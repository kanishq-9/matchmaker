```python
def calculate_match(user1, user2):
    score = 0
    total_checks = 0

    # Gender (prefer opposite)
    total_checks += 1
    if user1['gender'] != user2['gender']:
        score += 1

    # Age closeness (within 5 years)
    total_checks += 3
    if abs(user1['age'] - user2['age']) <= 5:
        score += 3
    elif abs(user1['age'] - user2['age']) <= 5:
        score += 2
    else:
        score+=1

    # Religion
    total_checks += 1
    if user1['religion'] == user2['religion']:
        score += 1

    # Caste
    total_checks += 1
    if user1['caste'] == user2['caste']:
        score += 1

    # Mother tongue
    total_checks += 3
    if user1['mother_tongue'] == user2['mother_tongue']:
        score += 3
    else:
        score += 1

    # Education
    total_checks += 5
    if user1['education'] == user2['education']:
        score += 5
    else:
        score += 1

    # Occupation
    total_checks += 2
    if user1['occupation'] == user2['occupation']:
        score += 2
    else:
        score += 1

    # Location (same city)
    total_checks += 2
    if user1['location'] == user2['location']:
        score += 2
    else:
        score += 1

    # Salary closeness (within 30%)
    total_checks += 2
    if abs(user1['salary_inr'] - user2['salary_inr']) / max(user1['salary_inr'], 1) <= 0.3:
        score += 2
    else:
        score += 1

    # Hobbies overlap
    hobbies1 = set(str(user1['hobbies']).split(","))
    hobbies2 = set(str(user2['hobbies']).split(","))
    common_hobbies = hobbies1 & hobbies2
    score += len(common_hobbies)   # add 1 for each hobby matched
    total_checks += max(len(hobbies1), len(hobbies2)) 

    # Convert to percentage
    percentage = round((score / total_checks) * 100, 2)
    return percentage


def find_matches(target_id):
    target = df[df['id'] == target_id].iloc[0]
    results = []

    for index, row in df.iterrows():
        if row['id'] == target_id:
            continue
        match_percent = calculate_match(target, row)
        results.append({
            "id": row['id'],
            "full_name": row['full_name'],
            "match_percentage": match_percent
        })

    matches = pd.DataFrame(results)
    matches = matches.sort_values(by="match_percentage", ascending=False)
    return matches


# Example usage
result = find_matches(1)
print(result.head(20))
```
